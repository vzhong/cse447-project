```python
import os
import string
import random
from argparse import ArgumentParser, ArgumentDefaultsHelpFormatter
```

```python
trainPath = r'../data/train'
files = os.listdir(trainPath)
for fileName in files:
    f = open(os.path.join(trainPath, fileName), 'r', encoding='utf-16')
    lines = f.read().split("\n")
    out = open(fileName, "w", encoding='utf-16')
    print(lines)
    for i in range(10):
        print(lines[i], file=out)
    out.close()
    f.close()

```

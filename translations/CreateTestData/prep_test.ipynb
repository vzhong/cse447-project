```python
import os
```

```python
path = r'test'
files = os.listdir(path)
lines = []
for f in files:
    f_ = open(os.path.join(path, f), "r", encoding='utf-16')
    curLines = f_.read().split('\n')
    curLines.pop()
    lines += curLines
    f_.close()
```

```python
path = r'test'
f = 'entest.corpus'
lines = []
f_ = open(os.path.join(path, f), "r", encoding='utf-16')
curLines = f_.read().split('\n')
curLines.pop()
lines += curLines
f_.close()
```

```python
prefix = open("prefix.txt", "w", encoding='utf-16')
answers = open("answers.txt", "w", encoding='utf-16')
```

```python
for line in lines:
    for i in range(1, len(line)):
        prefix.write(line[:i] + "\n")
        answers.write(line[i] + "\n" )
prefix.close()
answers.close()
```
